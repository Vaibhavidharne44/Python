# Finding the mean value of that column

```python
mean_value = data["normalized-losses"].astype("float").mean(axis=0)
print(mean_value)
```

```python
#aother col
mean_valueofhorsepower = data["horsepower"].astype("float").mean(axis=0)
print(mean_valueofhorsepower)
```

# Replacing mean values to null values for that column

```python
data["normalized-losses"].replace(np.nan,mean_value,inplace=True)
data.head()
```

```python
data["horsepower"].replace(np.nan,mean_valueofhorsepower,inplace=True)
data.head()
```
